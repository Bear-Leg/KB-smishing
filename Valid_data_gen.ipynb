```
nouns_all_token = []
morphs_all_token = []

with open('/content/gdrive/My Drive/KB_Smishing공모전/etc/mecab_nouns_num_all.pickle', 'rb') as f:
  nouns_all_token = pickle.load(f)

with open('/content/gdrive/My Drive/KB_Smishing공모전/etc/mecab_morphs_num_all.pickle', 'rb') as f:
  morphs_all_token = pickle.load(f)
```

```
print(len(nouns_all_token))
print(len(morphs_all_token))
print(morphs_all_token.head(5))
```

```
mecab_morphs_tokens = list(morphs_all_token['text'])
mecab_nouns_tokens = list(nouns_all_token['text'])
```

```
print(mecab_morphs_tokens[:5])
print(len(mecab_morphs_tokens))
print(mecab_nouns_tokens[:5])
print(len(mecab_nouns_tokens))
```

```
frequency_morphs_token_data = {}

for i in mecab_morphs_tokens:
  for j in i:
    if j not in frequency_morphs_token_data:
      frequency_morphs_token_data[j] = 1
    else:
      frequency_morphs_token_data[j] += 1
      
print(frequency_morphs_token_data)
print(len(frequency_morphs_token_data))
```

```
import operator

sdict= sorted(frequency_morphs_token_data.items(), key=operator.itemgetter(1))
print(sdict)
print(len(sdict))
for i in sdict:
  print(i)
```

```
# 빈도 n개 이하는 데이터는 거름
for i in list(frequency_morphs_token_data.keys()):
  if frequency_morphs_token_data[i] <= 6:
    del frequency_morphs_token_data[i]
print(len(frequency_morphs_token_data))
print(frequency_morphs_token_data)
```

```
frequency_nouns_token_data = {}

for i in mecab_nouns_tokens:
  for j in i:
    if j not in frequency_nouns_token_data:
      frequency_nouns_token_data[j] = 1
    else:
      frequency_nouns_token_data[j] += 1
print(frequency_nouns_token_data)
print(len(frequency_nouns_token_data))
```

```
sdict= sorted(frequency_nouns_token_data.items(), key=operator.itemgetter(1))
print(sdict)
print(len(sdict))
```

```
for i in list(frequency_nouns_token_data.keys()):
  if frequency_nouns_token_data[i] <= 2:
    del frequency_nouns_token_data[i]

print(len(frequency_nouns_token_data))
try:
  del frequency_nouns_token_data['ㄱ']
  del frequency_nouns_token_data['ㄴ']
  del frequency_nouns_token_data['ㄷ']
  del frequency_nouns_token_data['ㅁ']
  del frequency_nouns_token_data['ㅂ']
  del frequency_nouns_token_data['ㅅ']
  del frequency_nouns_token_data['ㅈ']
  del frequency_nouns_token_data['ㅣ']
except:
  print("DD")
print(frequency_nouns_token_data)
```

```
with open('/content/gdrive/My Drive/KB_Smishing공모전/etc/frequency_morphs_token_data.pickle', 'wb') as f:  
  pickle.dump(frequency_morphs_token_data, f, pickle.HIGHEST_PROTOCOL)

with open('/content/gdrive/My Drive/KB_Smishing공모전/etc/frequency_nouns_token_data.pickle', 'wb') as f:  
  pickle.dump(frequency_nouns_token_data, f, pickle.HIGHEST_PROTOCOL)
```
